# 牛场地区分割

```python
HERD=herd1=herd.copy()
```

```python
herd1['HERD']=list(map(lambda s:str(s),herd['HERD']))
```

```python
HERD['Region'] = HERD['HERD'].apply(lambda x: 'BJ' if x.startswith('11') else x)
HERD['Region'] = HERD['Region'].apply(lambda x: 'SH' if x.startswith('31') else x)
HERD['Region'] = HERD['Region'].apply(lambda x: 'NX' if x.startswith('64') else x)
HERD['Region'] = HERD['Region'].apply(lambda x: 'other' if len(x)>2 else x)
```
