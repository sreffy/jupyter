```python
wcc=pd.read_csv("C:\\Users\\mis\\Desktop\\wc\\data 1.csv",sep=",",encoding='gbk',low_memory=False)
```

```python
wcc1=wcc[wcc["胎组"]==1]
wcc2=wcc1[wcc1["是否稀释"]=='是']
wcc3=wcc1[wcc1["是否稀释"]=='否']

wcd1=wcc[wcc["胎组"]==2]
wcd2=wcd1[wcd1["是否稀释"]=='是']
wcd3=wcd1[wcd1["是否稀释"]=='否']

wce1=wcc[wcc["胎组"]==3]
wce2=wce1[wce1["是否稀释"]=='是']
wce3=wce1[wce1["是否稀释"]=='否']
```

```python
wcy1=wcc1.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')
wcy2=wcc2.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')
wcy3=wcc3.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')    
```

```python
wcy4=wcd1.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')
wcy5=wcd2.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')
wcy6=wcd3.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')    
```

```python
wcy7=wce1.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')
wcy8=wce2.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')
wcy9=wce3.pivot_table(index='DIM',columns='时间',values='体细胞数1000/mL',aggfunc='mean')    
```

```python
fig = plt.figure(figsize=(30,20))  
ax1 = fig.add_subplot(331)  
ax1.plot(wcy1)
plt.title('一胎牛')
plt.legend(wcy1)
ax2 = fig.add_subplot(334)  
ax2.plot(wcy2)
plt.title('稀释')
plt.legend(wcy2)
ax3 = fig.add_subplot(337)  
ax3.plot(wcy3)
plt.title('未稀释')
plt.legend(wcy3)
ax4 = fig.add_subplot(332)  
ax4.plot(wcy4)
plt.title('二胎牛')
plt.legend(wcy4)
ax5 = fig.add_subplot(335)  
ax5.plot(wcy5)
plt.title('稀释')
plt.legend(wcy5)
ax6 = fig.add_subplot(338)  
ax6.plot(wcy6)
plt.title('未稀释')
plt.legend(wcy6)
ax4 = fig.add_subplot(333)  
ax4.plot(wcy7)
plt.title('三胎及以上')
plt.legend(wcy7)
ax5 = fig.add_subplot(336)  
ax5.plot(wcy8)
plt.title('稀释')
plt.legend(wcy8)
ax6 = fig.add_subplot(339)  
ax6.plot(wcy9)
plt.title('未稀释')
plt.legend(wcy9)
plt.suptitle('按胎次分组')
plt.savefig("F:\\Python\\fig3.png",bbox_inches='tight',dpi=600)
```
